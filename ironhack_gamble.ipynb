```python
data = [account, customer, product, betting1, betting2, betting_summary, school, student]
data_title = ["account", "customer", "product", "betting1", "betting2", "betting_summary", "school", "student"]

for j in range(len(data)):
    table = data[j]
    columns = table.columns
    parameter = ", ".join(list(columns))
    param_2 = ("%s, " * len(columns))[:-2]

    query = f"INSERT INTO {data_title[j]}({parameter}) VALUES ({param_2})"
    print(f"Query: {query}")
    
    for i in range(len(data[j])):
        row = tuple(table.iloc[i])
        # print(f"Row: {row}")
        
        try:
            cursor.execute(query, row)
        except Exception as e:
            print(f"Error executing query: {e}")
            cnx.rollback()  # Rollback any uncommitted transactions
            break
```

```python
try:
    cnx.commit()
except Exception as e:
    print(f"Error committing transaction: {e}")
    cnx.rollback()
```
